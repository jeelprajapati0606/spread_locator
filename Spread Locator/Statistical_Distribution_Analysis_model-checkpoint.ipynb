# Spread Locator: A Statistical Distribution Analysis Model :

# Import Libraries :

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats
from scipy.stats import bernoulli, binom, poisson, lognorm, powerlaw, zscore, norm

from scipy.stats import boxcox
import statsmodels.api as sm
```

## Import Data :

```python
df = pd.read_csv("Speed_Locator_Data.csv")
df.head()
```

## 1. Fit the data to Bernoulli and Binomial distributions (transaction occurrence & weekly count).

```python
# Bernoulli Distribution (transaction occurrence) 

# Success = 1, Fail = 0
df["transaction_occurrence"] = df["transaction_status"].map({
    "Success": 1,
    "Fail": 0
})

sns.countplot(x=df["transaction_occurrence"])
plt.title("Bernoulli Distribution - Transaction Occurrence")
plt.xlabel("Transaction (0=Fail, 1=Success)")
plt.ylabel("Count")
plt.show()


```

### Interpretation :
- The graph has two outcomes: Success (1) and Fail (0).
- Failed transactions are more than successful transactions.

```python
# Binomial Distribution (Weekly Transaction Count)

sns.histplot(df["transaction_count"], bins=10)

plt.title("Binomial Distribution - Weekly Transaction Counts")
plt.xlabel("Number of Transactions per Week")
plt.ylabel("Frequency")

plt.show()
```

## Interpretation :
- Most transaction counts fall within a specific range.
- Higher transaction counts occur less frequently.
- The distribution shows the weekly transaction pattern.

# 2. Fit the data to Poisson distribution (number of transactions per day).


```python
Daily_Transaction = df.groupby("transaction_date").size()

lam = Daily_Transaction.mean()
print("Lambda: ",lam)

x = np.arange(0,Daily_Transaction.max()+3)

pmf = poisson.pmf(x,lam)

plt.figure(figsize =(7,4))
plt.bar(x,pmf)
plt.title("Poisson Distribution")
plt.xlabel("Transactions Per Day")
plt.ylabel("Probability")
plt.show()
```

## Interpretation :

- Most probable value: The peak occurs at 6 transactions/day, meaning this is the expected average (λ) and the most likely outcome.

- Probability spread: Values close to 6 (like 5 or 7) also have relatively high probabilities, while extreme values (0 or 14) are rare.

- Skewed shape: The distribution is slightly right-skewed, which is typical when λ is moderate. It’s not perfectly symmetric like a normal distribution.

- Event modeling: This distribution is useful for modeling random, independent events over time — here, daily transactions.

# 3. Model transaction amounts using Log-Normal ..

```python
amount = df["transaction_amount"]

shape,loc,scale = lognorm.fit(amount, floc=0)

print("Shape",shape)
print("Scale",scale)

plt.figure(figsize=(8,5))

sns.histplot(amount, bins=30, stat='density')

x = np.linspace(amount.min(), amount.max(),300)

pdf = lognorm.pdf(x,shape,loc,scale)

plt.plot(x,pdf,'r',linewidth=2)

plt.title("log normal fit")
plt.show()
```

## Interpretation :

- Right-skewed distribution: Most transaction amounts are concentrated at lower values (0–5000), while fewer transactions occur at higher amounts, creating a long right tail.

- Log-normal fit: The red curve shows that a log-normal distribution models the data well, capturing the skewness typical of financial transaction amounts.

- Density meaning: The y-axis represents density, not raw counts, meaning the curve is scaled to show relative likelihood rather than absolute frequency.

# 4. Generate and interpret a Q-Q Plot to test normality.

```python
amount = df["transaction_amount"]

plt.figure(figsize=(6,6))

stats.probplot(amount, dist="norm",plot=plt)
plt.title("Q-Q Plot")
plt.show()

# Shapiro-Wilk Test

stat, pvalue = stats.shapiro(amount)

print("Shapiro Statistic:", stat)
print("P-value:", pvalue)

if pvalue > 0.05:
    print("Data appears Normally Distributed.")
else:
    print("Data is NOT Normally Distributed.")
```

## Interpretation :

- Strong deviation: The blue data points stray far from the red line, especially at the tails, showing the dataset does not follow a normal distribution.

- Statistical evidence: The Shapiro statistic (≈0.736) and extremely small p-value (~1.9e-18) confirm statistically that the data is not normal.

- Modeling impact: Since normality is rejected, alternative distributions (like log-normal or power-law) are better suited for analyzing transaction amounts.

# 5. Apply Box-Cox Transform to stabilize variance.

```python
amount = df["transaction_amount"]

# Box-Cox requires positive values

positive_amount = amount + 1

transformed, lambda_value = stats.boxcox(positive_amount)
print("Optimal Lambda ", lambda_value )

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.hist(amount, bins=30)
plt.title("Original Data")

plt.subplot(1,2,2)
plt.hist(transformed, bins=30, color="green")
plt.title("After Box-Cox")

plt.show()
```

## Interpretation :

- Original skewness: The left histogram shows the raw data is heavily right-skewed, with many small values and a few very large outliers.

- Transformation effect: After applying Box-Cox, the right histogram looks more symmetric and closer to a normal distribution.

- Optimal lambda: The reported λ (≈ -0.18) is the best parameter that reduces skewness, making the data suitable for statistical modeling.

# 6. Calculate Z-scores for transaction amounts and compute probability of transactions exceeding ₹5000.

```python
amount = df["transaction_amount"]

df ["Z_score"]= zscore(amount)

print(df[["transaction_amount", "Z_score"]].head())

# Probability of transaction amount > ₹5000

mean = amount.mean()

std = amount.std()

probability = 1- norm.cdf(5000, loc =mean, scale =std)

print("Probability of Transaction Amount > ₹5000:", probability)
```

## Interpretation :

- Z-scores meaning: They show how far each transaction is from the average — most are close to normal, while 6383.78 is higher than average.

- Normal vs unusual: Transactions around 2700–4100 are typical, while values above 5000 are less common.

- Probability insight : About 20.5% of transactions exceed ₹5000, highlighting the chance of larger amounts.

# 7. Plot and interpret PDF and CDF for transaction amounts.

```python
amount = df["transaction_amount"]

mean = amount.mean()
std = amount.std()

x = np.linspace(amount.min(), amount.max(), 500)


# PDF

pdf = norm.pdf(x, mean, std)

plt.figure(figsize =(8,5))
plt.plot(x,pdf)
plt.title("Probability Density Function (PDF)")
plt.xlabel("Transaction Amount")

plt.ylabel("Density")
plt.grid(True)
plt.show()

# CDF

cdf = norm.cdf(x, mean, std)

plt.figure(figsize=(8,5))
plt.plot(x, cdf, color="green")
plt.title("Cumulative Distribution Function (CDF)")
plt.xlabel("Transaction Amount")
plt.ylabel("Cumulative Probability")
plt.grid(True)
plt.show()
```

## Interpretation :

- PDF insight: The blue curve shows most transactions cluster around ~4000, with fewer very small or very large amounts. This highlights the peak of typical transaction values.

- CDF meaning: The green curve accumulates probability as transaction amounts increase. It rises steeply between 2000–7000, meaning most transactions fall in this range.

- Practical use: Together, PDF and CDF help in understanding both the likelihood of specific transaction amounts and the overall probability distribution, useful for forecasting and risk analysis.

# 8. Conclude which distribution best fits the dataset and justify insights for decision-making.

```python
amount = df["transaction_amount"]

# Log-Normal Fit

log_shape, log_loc, log_scale = lognorm.fit(amount, floc=0)

log_lh = lognorm.logpdf(amount, log_shape, log_loc, log_scale).sum()

# Power Law Fit

pow_shape, pow_loc, pow_scale = powerlaw.fit(amount)
pow_lh = powerlaw.logpdf(amount, pow_shape, pow_loc, pow_scale).sum()

print("Log-Normal Log-Likelihood:", log_lh)
print("Power Law Log-Likelihood:", pow_lh)

if log_lh > pow_lh:
    print("\nBest Fit: Log-Normal Distribution")
else:
    print("\nBest Fit: Power Law Distribution")
```

## Interpretation :

- Log-likelihood comparison: The Log-Normal model has a higher (less negative) log-likelihood (-1908) compared to the Power Law (-2056), meaning it fits the data better.

- Best fit model: Since higher log-likelihood indicates better fit, the Log-Normal distribution is the preferred choice for modeling transaction amounts.

- Practical implication: Using Log-Normal helps capture the skewed nature of financial transactions more accurately, improving analysis and predictions.
